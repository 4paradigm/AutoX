```python
from autox import AutoX
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
# 选择数据集
data_name = 'elo'
path = f'../data/{data_name}'
```

```python
relations = [
    {
            "related_to_main_table": "true", # 是否为和主表的关系
            "left_entity": "train.csv",  # 左表名字
            "left_on": ["card_id"],  # 左表拼表键
            "right_entity": "historical_transactions.csv",  # 右表名字
            "right_on": ["card_id"], # 右表拼表键
            "type": "1-M" # 左表与右表的连接关系
        },
    {
            "related_to_main_table": "true", # 是否为和主表的关系
            "left_entity": "test.csv",  # 左表名字
            "left_on": ["card_id"],  # 左表拼表键
            "right_entity": "historical_transactions.csv",  # 右表名字
            "right_on": ["card_id"], # 右表拼表键
            "type": "1-M" # 左表与右表的连接关系
        },
    {
            "related_to_main_table": "true", # 是否为和主表的关系
            "left_entity": "train.csv",  # 左表名字
            "left_on": ["card_id"],  # 左表拼表键
            "right_entity": "new_merchant_transactions.csv",  # 右表名字
            "right_on": ["card_id"], # 右表拼表键
            "type": "1-M" # 左表与右表的连接关系
        },
    {
            "related_to_main_table": "true", # 是否为和主表的关系
            "left_entity": "test.csv",  # 左表名字
            "left_on": ["card_id"],  # 左表拼表键
            "right_entity": "new_merchant_transactions.csv",  # 右表名字
            "right_on": ["card_id"], # 右表拼表键
            "type": "1-M" # 左表与右表的连接关系
        }
]

autox = AutoX(target = 'target', train_name = 'train.csv', test_name = 'test.csv', 
               id = ['card_id'], path = path, relations = relations)
```

```python
sub = autox.get_submit()
```
